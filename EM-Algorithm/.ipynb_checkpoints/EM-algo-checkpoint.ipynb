```python
from sklearn.datasets.samples_generator import make_blobs
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import mixture
```

```python
X, y = make_blobs(n_samples=100, centers=2, n_features=1,random_state=23)
```

```python
x1, x2 = -10, 20
y1, y2 = -5, 5
plt.axis([x1, x2, y1, y2])
plt.plot(X, y, 'ro')
plt.show()
```

```python
data = pd.read_csv('unbalance.txt', header = None,sep=" ")
data.columns = ["X", "y"]
X = data.X
y = data.y

plt.plot(X, y , 'ro')
plt.show()
```

```python
gmm = mixture.GaussianMixture(n_components=2,covariance_type='full')
gmm.fit(data)
```

```python
gmm.bic(data)
```

```python
data_probability = gmm.predict_proba(data)
prob = pd.DataFrame(data_probability)
print prob
prob.columns = ["first", "zero"]
```

```python
data_1_X = []
data_1_y = []
data_0_X = []
data_0_y = []

for index,row in prob.iterrows():
    if(row['first']>row['zero']):
        data_1_X.append( data.iloc[index].X )
        data_1_y.append( data.iloc[index].y )
    else:
        data_0_X.append( data.iloc[index].X )
        data_0_y.append( data.iloc[index].y )
```

```python
plt.plot(data_1_X, data_1_y, 'ro'  )
plt.plot( data_0_X, data_0_y, 'bo')
plt.show()
```

```python
gmm1 = mixture.GaussianMixture(n_components=8,covariance_type='full')
gmm1.fit(data)
res = gmm1.predict(data)
```

```python
res_data = data.assign(cluster_num=pd.Series(res).values)
```

```python
res = res_data.groupby('cluster_num')
```

```python
res
```

```python
colors = ['bo','go','ro','co','mo','yo','ko','wo']
```

```python
for i in res.groups:
    sample_X = []
    sample_Y = []
    color = colors[i]
    for j in res.groups[i]:
        sample_X.append( data.iloc[j].X )
        sample_Y.append( data.iloc[j].y )
    plt.plot(sample_X, sample_Y, color )
plt.show()

        
```
